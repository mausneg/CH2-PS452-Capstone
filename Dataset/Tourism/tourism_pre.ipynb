```python
import pandas as pd
import json
```

```python
df_tourism = pd.read_csv('raw/tourims.csv')
df_tourism.head()
```

```python
df_tourism.drop(['Column1','_1'], axis=1, inplace=True)
```

```python
df_tourism['Coordinate'] = df_tourism['Coordinate'].apply(lambda x: x.replace("{'lat': ",'').replace(" 'lng':",'').replace("}",'').replace(", ",','))
df_tourism[['Lat','Long']] = df_tourism['Coordinate'].str.split(',',expand=True).astype(float)
df_tourism.drop(['Coordinate'], axis=1, inplace=True)
```

```python
df_tourism['Rating'] = df_tourism['Rating'].apply(lambda x: x/10 if x > 10 else x).astype(float)
```

```python
df_tourism.head()
```

```python
df_tourism.to_csv('tourism.csv', index=False)
```
