```python
import pandas as pd
```

```python
df_jakarta = pd.read_csv('raw/db_hotel_jakarta.csv')
df_jatim = pd.read_csv('raw/db_hotel_jatim.csv')
df_jateng = pd.read_csv('raw/db_hotel_jateng.csv')
df_jabar = pd.read_csv('raw/db_hotel_jabar.csv')
df_diy = pd.read_csv('raw/db_hotel_diy.csv')
df_bali = pd.read_csv('raw/db_hotel_bali.csv')
```

```python
resort_id = []
resort_image = []
resort_name = []
resort_type = []
resort_city = []
resort_address = []
resort_rating = []
resort_min_price = []
resort_max_price = []
resort_lat = []
resort_long = []

for df in [df_bali, df_diy, df_jabar, df_jakarta, df_jateng, df_jatim]:
    resort_id.append(df.columns[0])
    resort_image.append(df.columns[1])
    resort_name.append(df.columns[2])
    resort_type.append(df.columns[3])
    resort_city.append(df.columns[4])
    resort_address.append(df.columns[5])
    resort_rating.append(df.columns[6])
    resort_min_price.append(df.columns[7])
    resort_max_price.append(df.columns[8])
    resort_lat.append(df.columns[9])
    resort_long.append(df.columns[10])
    resort_id.extend(df.iloc[:, 0])
    resort_image.extend(df.iloc[:, 1])
    resort_name.extend(df.iloc[:, 2])
    resort_type.extend(df.iloc[:, 3])
    resort_city.extend(df.iloc[:, 4])
    resort_address.extend(df.iloc[:, 5])
    resort_rating.extend(df.iloc[:, 6])
    resort_min_price.extend(df.iloc[:, 7])
    resort_max_price.extend(df.iloc[:, 8])
    resort_lat.extend(df.iloc[:, 9])
    resort_long.extend(df.iloc[:, 10])
```

```python
df_resorts = pd.DataFrame({
    'Resort_Id' : resort_id,
    'Resort_Image' : resort_image,
    'Resort_Name' : resort_name,
    'Type' : resort_type,
    'City' : resort_city,
    'Address' : resort_address,
    'Rating' : resort_rating,
    'Min_Price' : resort_min_price,
    'Max_Price' : resort_max_price,
    'Lat' : resort_lat,
    'Long' : resort_long 
})
```

```python
df_resorts['Resort_Id'] = df_resorts['Resort_Id'].astype(int)
df_resorts.sort_values(by=['Resort_Id'], inplace=True)
df_resorts.reset_index(drop=True, inplace=True)
df_resorts
```

```python
df_resorts.dropna(inplace=True)
```

```python
df_resorts.to_csv('resorts.csv', index=False)
```
