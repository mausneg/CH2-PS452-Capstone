```python
import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df_tourism_rating = pd.read_csv('../../../Dataset/Tourism Rating/raw/tourism_rating.csv')
df_tourism_rating
```

```python
df_tourism =pd.read_csv('../../../Dataset/Tourism/tourism.csv')
df_tourism
```

```python
df_rating = pd.merge(df_tourism_rating, df_tourism, on='Place_Id',how='left')
df_rating
```

```python
drop_columns = df_rating.columns[7:]
df_rating = df_rating.drop(drop_columns, axis=1)
```

```python
df_rating['Category'] = df_rating['Category'].apply(lambda x: x.replace(',', ' '))
df_rating['Tags'] = df_rating['Description'] + ' ' + df_rating['Category'] + ' ' + df_rating['City']
df_rating.drop(['Description', 'Category', 'City'], axis=1, inplace=True)
df_rating
```

```python
stopwords = pd.read_csv('../../../Dataset/Stop Words/stopwordbahasa.csv')
stopwords_list = stopwords.iloc[:,0].tolist()
```

```python
cv = CountVectorizer(max_features = 5000, lowercase=True, stop_words=stopwords_list)
user_preferences_feature = cv.fit_transform(df_rating['Tags']).toarray()
user_preferences_feature.shape
```
